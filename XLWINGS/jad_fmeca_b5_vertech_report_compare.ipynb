### Check duplicates

Vertech says a number of anomalies were duplicates. The duplicates are listed here next to each other in the list below>

```python
dups = ["4SWBT-LBHD- CR-286",
"4SWBT-LBHD- CR-609",
"4SWBT-LBHD- CR-287",
"4SWBT-LBHD- CR-608",
"4SWBT-LBHD- CR-288",
"4SWBT-LBHD- CR-606",
"4SWBT-LBHD- CR-289",
"4SWBT-LBHD- CR-604",
"4SWBT-LBHD- CR-295",
"4SWBT-LBHD- CR-603",
"4SWBT-LBHD- CR-297",
"4SWBT-LBHD- CR-605",
"4SWBT-LBHD- CR-298",
"4SWBT-LBHD- CR-602",
"4SWBT-LBHD- CR-299",
"4SWBT-LBHD- CR-601",
"4SWBT-LBHD- CR-1379",
"4SWBT-LBHD- CR-1378",
]

# -- Create a dictionary that maps each item to its position

dups_dict = {item: i for i, item in enumerate(dups)}

# -- Create a new column that contains the position of each item in the dataframe
df['dup_position'] = df['Anomaly ID'].apply(lambda x: dups_dict.get(x,-1))

# -- Create a df with duplicate objects and then sort the values based on "dup_position"
dup_df = df[df['Anomaly ID'].isin(dups)].sort_values('dup_position')
```
